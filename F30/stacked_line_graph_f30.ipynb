```python
# libraries
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# set seaborn style
sns.set_theme(style='white')
 
```

```python
# Data
actigraph = pd.read_csv('155_50hours.csv')
x = range(1,51)
y1 = [actigraph['% Sedentary']]
y2 = [actigraph['% Light']]
y3 = [actigraph['% Moderate']]
y4 = [actigraph['% Vigorous']]
```

```python
print('Default')
plt.stackplot(x,y1, y2, y3, y4, labels=['Sedentary','Light','Moderate','Vigorous'])
plt.show()
print('Resized')
f = plt.figure()
f.set_figwidth(10)
f.set_figheight(4)

plt.stackplot(x,y1, y2, y3, y4, labels=['Sedentary','Light','Moderate','Vigorous'])
plt.savefig('test.png')#fig = plt.stackplot(x,y1, y2, y3, y4, labels=['Sedentary','Light','Moderate','Vigorous'])
#fig.set_figwidth(1)
```

```python
# Basic stacked area chart.
fig = plt.stackplot(x,y1, y2, y3, y4, labels=['Sedentary','Light','Moderate','Vigorous'])
plt.legend()
plt.figure(figsize=(5,1000))
```

```python
# Data
actigraph = pd.read_csv('percent_155.csv')

x = range(0,99)
y1 = [actigraph['% in Sedentary']]
y2 = [actigraph['% in Light']]
y3 = [actigraph['% in MVPA']]

fig = plt.stackplot(x, y1, y2, y3, labels=['Sedentary','Light','Moderate-to-Vigorous'])
plt.legend()
print('Resized')
f = plt.figure()
f.set_figwidth(10)
f.set_figheight(4)

pal = sns.color_palette("vlag")

plt.stackplot(x,y1, y2, y3, labels=['Sedentary','Light','Moderate-to-Vigorous'], colors = pal)
plt.legend()
plt.title('13 Year-Old Male')
plt.xlabel('Hours')
plt.ylabel('Percent')
```
