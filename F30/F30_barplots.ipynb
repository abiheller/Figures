```python
actigraph = pd.read_csv('percent_150.csv')
Counts = actigraph["Counts"]
Sedentary = actigraph["% in Sedentary"]
Light = actigraph["% in Light"]
Moderate = actigraph["% in Moderate"]
Vigorous = actigraph["% in Vigorous"]
VeryVigorous = actigraph["% in Very Vigorous"]

width = 1    # the width of the bars: can also be len(x) sequence


fig, ax = plt.subplots()

ax.bar(Counts, Sedentary, width, label='Sedentary', color = '#ebd3d5')
ax.bar(Counts, Light, width, bottom=Sedentary,
       label='Light', color = '#f799a1')
ax.bar(Counts, Moderate, width, bottom=Light+Sedentary, label = 'Moderate', color = '#fc6a76')
ax.bar(Counts, Vigorous, width, bottom=Moderate+Light+Sedentary, label = 'Vigorous', color = '#fc0317')
ax.bar(Counts, VeryVigorous, width, bottom=Vigorous+Moderate+Light+Sedentary, label = 'Very Vigorous', color = '#7a000a')

params = {'mathtext.default': 'regular' }          
plt.rcParams.update(params)

ax.set_ylabel('Percent', size = 16)
ax.set_title('9 Year-Old Boy, Estimated $VO_2$ Max = 43.3 mL/kg/min', size = 18)

ax.legend(bbox_to_anchor=(1, 1), loc='upper left', borderaxespad=0)

plt.xticks(color = 'w')
handles, labels = ax.get_legend_handles_labels()
ax.legend(reversed(handles), reversed(labels), title='Movement Classification', bbox_to_anchor=(1.5,1))
plt.yticks(color = 'w')
f = fig
f.set_figwidth(10)
f.set_figheight(1.5)

plt.show()

```
